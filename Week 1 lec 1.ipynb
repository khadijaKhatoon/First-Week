
##  The Personal Assistant 

Create a simple program that acts as a basic personal assistant by asking for user input and performing simple operations.

```python
## 1. Ask the user for their name and store it in a variable
name = input('what is your name:')  ## Ask the user for their name 
print(f"Hello, {name}! Nice to meet you.")  ## Display a greeting with the user's name
```

```python
## 2. Ask the user for their birth year (as an integer) and store it.
DOB = input('what year where you born:')  ##  Ask the user for their DOB 
print(f"Your DOB is {DOB}")               ##  Display a greeting with the user's DOB
```

```python
## 3. Ask the user for their current age (as an integer) and store it.
Age = input('What is your current age?')   ##  Ask the user for their Age
print(f"Your age is {Age}")                ##  Display a greeting with the user's Age
```

```python
## 4. Calculate and print the year they will turn 100 years old. (Hint: 100 - age + current_year).
name = input("What is your name? ")  ## Ask for user name
age = int(input("How old are you? "))  ## Ask for user age

# Get current year (you can either hardcode it or use datetime module)
# Option 1: Hardcode the current year
current_year = 2026

# Calculate the year they will turn 100
year_turn_100 = current_year + (100 - age)

# Print the result
print(f"Hello {name}! You will turn 100 years old in the year {year_turn_100}.")
```

```python
## 5. Ask the user for a lucky number (as a float).
No=input("Enter your lucky number?")  ## Ask for user lucky number
print(f"Your lucky number is {No}")   ##  Display a greeting with the user's lucky number
```

```python
## 6. Print a summary message that includes all the information you've collected. Use appropriate data types and typecasting.
print("...Personal Assistant Summary...")

print(f"Hello, {name}! Nice to meet you.")
print(f"Your DOB is {DOB}")
print(f"Your age is {Age}")
print(f"Hello {name}! You will turn 100 years old in the year {year_turn_100}.")
print(f"Your lucky number is {No}")
print("------------------------------")
```
