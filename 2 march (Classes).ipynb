```python
# Base Class
class Animal:
  def sound(self):
    print("Animal makes a sound.")

## Subclass Dog
class Dog(Animal):
 def sound(self):
    print("Dog says : woof")
class Cat(Animal):
 def sound(self):
    print("Cat says : Meow")
class Cow(Animal):
 def sound(self):
    print("Cow says : Moo")


dog = Dog()
cat = Cat()
cow = Cow()
 
dog.sound()
cat.sound()
cow.sound()

    
```

```python
class Vehicle:
    def move(self):
        print("Vehicle is moving.")
class car(Vehicle):
    def move(self):
        print("Car is driving on the road.")
class Bike(Vehicle):
    def move(self):
        print("Bike is riding on the street.")
class Bus(Vehicle):
    def move(self):
        print("Bus is transporting passengers.")

v1 = car()
v2 = Bike()
v3 = Bus()

Vehicle = [v1,v2,v3]
for v1 in Vehicle:
    v1.move()
```

```python
import math

# Base Class
class Shape:
    def area(self):
        print("Area calculation not defined")

# Subclass Rectangle
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):   # Overriding
        return self.length * self.width


# Subclass Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):   # Overriding
        return math.pi * self.radius ** 2


# Subclass Triangle
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):   # Overriding
        return 0.5 * self.base * self.height


# Creating Objects
r = Rectangle(5, 4)
c = Circle(3)
t = Triangle(6, 2)

# Storing in list (Polymorphism)
shapes = [r, c, t]

# Loop to calculate area
for shape in shapes:
    print("Area:", shape.area())
```

```python
class Shape:
    def area(self):
        print("Area not defined")

# Rectangle subclass
class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print("Rectangle Area:", self.length * self.width)

# Circle subclass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print("Circle Area:", math.pi * self.radius * 2)

# Triangle subclass
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        print("Triangle Area:", 0.5 * self.base * self.height)

r = Rectangle(5, 4)
c = Circle(3)
t = Triangle(6, 2)

r.area()
c.area()
t.area()
```

```python
# Base class
class Employee:
    def calculate_salary(self):
        print("Salary not defined")

# Full-time employee
class FullTimeEmployee(Employee):
    def __init__(self, salary):
        self.salary = salary

    def calculate_salary(self):
        print("Full-Time Salary:", self.salary)

# Part-time employee
class PartTimeEmployee(Employee):
    def __init__(self, hourly_rate, hours):
        self.hourly_rate = hourly_rate
        self.hours = hours

    def calculate_salary(self):
        if self.hourly_rate >= 0 and self.hours >= 0:
            total = self.hourly_rate * self.hours
            print("Part-Time Salary:", total)
        else:
            print("Invalid input for Part-Time employee")

# Intern
class Intern(Employee):
    def __init__(self, stipend):
        self.stipend = stipend

    def calculate_salary(self):
        print("Intern Stipend:", self.stipend)


# Runtime polymorphism
employees = [
    FullTimeEmployee(50000),
    PartTimeEmployee(500, -30),  # Negative hours to test invalid input
    Intern(15000)
]

for emp in employees:
    emp.calculate_salary()
```
