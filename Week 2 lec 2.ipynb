## Shape Calculator with Magic Methods

```python

# STEP 1: Create class Rectangle with width and height
class Rectangle:
    def __init__(self, width, height):
        self.width = width    # Initialize width attribute
        self.height = height  # Initialize height attribute
    


    # STEP 2: Implement __str__ method for string representation
    def __str__(self):
        return f"Rectangle(width={self.width}, height={self.height})"
    


    # STEP 3: Implement area() method returning width * height
    def area(self):
        return self.width * self.height
    


    # STEP 4: Overload + operator using __add__ to return new rectangle
    def __add__(self, other):
        # Create and return a new rectangle with widths and heights added
        return Rectangle(self.width + other.width, self.height + other.height)
    


    # STEP 5: Overload == operator using __eq__ to compare areas
    def __eq__(self, other):
        # Compare rectangles based on their areas
        return self.area() == other.area()



# STEP 6: Create rectangles rect1(4,5) and rect2(2,3)
rect1 = Rectangle(6, 5)
rect2 = Rectangle(9, 3)



# STEP 7: Print rectangles and areas
print(f"Rectangle 1: {rect1}")  # __str__ method automatically called
print(f"Rectangle 2: {rect2}")  # __str__ method automatically called
print(f"Area of rect1: {rect1.area()}")
print(f"Area of rect2: {rect2.area()}")
print()



# STEP 8: Create rect3 = rect1 + rect2 (using overloaded + operator)
rect3 = rect1 + rect2  # This calls the __add__ method
print(f"rect3 (rect1 + rect2): {rect3}")
print(f"Area of rect3: {rect3.area()}")
print()



# STEP 9: Compare rectangles using overloaded == operator
print(f"Are rect1 and rect2 equal? {rect1 == rect2}")  # Compares areas

# Create rect4 with same dimensions as rect2 for comparison
rect4 = Rectangle(5, 3)
print(f"Are rect2 and rect4 equal? {rect2 == rect4}")  # Should be True (same dimensions)

# Additional comparisons
rect5 = Rectangle(2, 4)  # Area = 12 (same as rect3)
print(f"\nrect5: {rect5}")
print(f"Area of rect5: {rect5.area()}")
print(f"Are rect3 and rect5 equal? {rect3 == rect5}")  # Compare areas (both 24)
```
