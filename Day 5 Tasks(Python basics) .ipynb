## Tasks: If, Elif, Else

```python
# 1. Check if a number is positive, negative, or zero
x=56
if(x>34):
   print("Number is positive.")
elif(x<34):
   print("Number is negative.")
else:
   print("NUmber is Zero")


# 2. Determine if a person is eligible to vote (age >= 18)
age=int(input("Enter your age:"))
if(age>=18):
  print("you are eligible to vote.")
else:
   print("you are not eligible to vote.")



# 3. Print grade based on score (A/B/C/D/F)
score= int(input("Enter your score:"))
if(score>=90):
   print("Grade A.")
elif(score>=80):
   print("Grade B.")
elif(score>=70):
   print("Grade C.")
elif(score>=60):
   print("Grade D.")
else:
   print("Grade F.")


# 4. Check if a number is even or odd
num = int(input("Enter any number:"))
if(num%2 ==0):
   print("Your number is Even.")
else:
   print("Your number is odd.")

# 5. Find the largest of two numbers
a=45
k=67
if(a>=k):
   print("Largest number is 45.")
else:
   print("Largest number is 67.")


# 6. Check if temperature indicates Hot/Warm/Cold
temperature = float(input("Enter temperature: "))
if temperature >= 30:
    print("Hot")
elif temperature >= 20:
    print("Warm")
else:
    print("Cold")


# 7. Determine if a character is a vowel or consonant
ch = input("Enter any chracter:")
if ch.lower in ['a', 'e', 'i', 'o' ,'u']:
  print("It is a vowel.")
else:
   print("It is a consonant.")

# 8. Check if a number is divisible by both 3 and 5
num = int(input("Enter any number:"))
if(num%3==0 and num%5==0):
   print("Number is divisible by both 3 and 5.")
else:
   print("Number is not divisilbe by both 3 and 5. ")

# 9. Determine ticket price based on age group
age = int(input("Enter your age:"))
if(age<5):
   print("Ticket is free.")
elif(age<12):
   print("Ticket price is 200 Rs.")
elif(age<40):
   print("Ticket price is 350 Rs.")
else:
   print("Ticket pice is 300 Rs. (Senior citizen discount).")
```

## Tasks: Break, Continue, Pass

```python
# 1. Print numbers until 5 using break
for i in range(10):
  if i==5:
    break
  print(i)
```

```python
# 2. Skip number 3 using continue
for i in range(10):
    if i==3:
        continue
    print(i)
```

```python
# 3. Loop with pass as placeholder
for i in range(10):
  if i==5:
    pass
  print(i)
```

```python
# 4. Stop loop when a negative number appears
for i in range(-1,7):
    if i<0:
        continue
    print(i)
```

```python
# 5. Skip even numbers in a loop
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
# 6. Break when sum exceeds 50
total=0
for i in range(20):
    total+=i
    if total > 50:
      break
    print("Final sum:", total)
```

```python
# 7. Continue for vowels only in a string
text="khadija"
for ch in text:
    if ch.lower() not in ['a', 'e','i','o','u']:
        continue
    print(ch)
```

```python
# 8. Infinite loop that exits with break
for i in range(10):   # large range
    print("Loop is running...")
    
    user = input("Type 'exit' to stop: ")
    
    if user == "exit":
        print("Loop stopped!")
        break
```

```python
# 9. Skip multiples of 4
for i in range(1 , 21):
    if i%4==0:
        continue
    print(i)
```

```python
# 10. Use pass inside an if condition
num =10
if(num>5):
  print("pass")
else:
 print("fail")
```

## Tasks: Nested Conditionals

```python
# 1. Check positive/negative then even/odd
num = int(input("Enter a number: "))

if num >= 0:
    print("Number is Positive")
else:
    print("Number is Negative")

if num % 2 == 0:
    print("It is Even")
else:
    print("It is Odd")
```

```python
# 2. Age check for driving and insurance category
age = int(input("Enter age: "))

if age >= 18:
    print("Eligible for Driving")
    
    if age < 25:
        print("Insurance: High Premium")
    else:
        print("Insurance: Normal Premium")
else:
    print("Not eligible for Driving")
```

```python
# 3. Score check for pass and distinction
score = int(input("Enter score: "))

if score >= 50:
    print("Pass")
    
    if score >= 80:
        print("Distinction")
else:
    print("Fail")
```

```python
# 4. Check number range then divisibility
num = int(input("Enter number: "))

if 1 <= num <= 100:
    print("Number in range")
    
    if num % 5 == 0:
        print("Divisible by 5")
else:
    print("Out of range")
```

```python
# 5. Weather check: temp and rain
temp = int(input("Enter temperature:"))
rain=input("It is raining?(yes/no):")

if(temp>30):
    print("Hot weather.")
    if rain == "yes":
        print("Humid and rainy")
else:
    print("Normal weather.")
```

```python
# 6. Salary check for tax slab
salary = int(input("Enter your salary:"))
if salary>500000:
    print("Tax aplicable.")
    if salary>100000:
        print("High tax slab.")
    else:
        print("Norml tax slab.")
else:
    print("No tax.")
```

```python
# 7. Check login: username then password
username = input("Enter username:")
password = input("Enter password:")
if username == "khadija":
    if password == "1234":
        print("Login Successful.")
    else:
        print("Wrong password.")
else:
    print("Invalid username.")
```

```python
# 8. Check triangle type by sides
a = int(input("Side 1: "))
b = int(input("Side 2: "))
c = int(input("Side 3: "))

if a == b == c:
    print("Equilateral Triangle")
elif a == b or b == c or a == c:
    print("Isosceles Triangle")
else:
    print("Scalene Triangle")
```

```python
# 9. Check marks then scholarship eligibility
marks = int(input("Enter marks: "))

if marks >= 60:
    print("Eligible for Admission")
    
    if marks >= 85:
        print("Eligible for Scholarship")
else:
    print("Not Eligible")
```
