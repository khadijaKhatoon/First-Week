```python
class Bike:
    def __init__(self, brand, model, price):
        # Private attributes
        self.__brand = brand
        self.__model = model
        self.__price = price

        # Protected attribute
        self._category = "Standard"

    # Getter Methods
    def get_brand(self):
        return self.__brand

    def get_price(self):
        return self.__price

    # Setter Method
    def set_price(self, new_price):
        self.__price = new_price

    # Method
    def calculate_discount(self):
        return self.__price * 0.05   # 5% discount

    def display_info(self):
        print("Brand:", self.__brand)
        print("Model:", self.__model)
        print("Price:", self.__price)
        print("Category:", self._category)


class SportBike(Bike):
    def __init__(self, brand, model, price, top_speed):
        # Inherit parent constructor
        super().__init__(brand, model, price)

        # Additional private attribute
        self.__top_speed = top_speed
        self._category = "Sports"

    # Method override (Polymorphism)
    def calculate_discount(self):
        return self.get_price() * 0.10   # 10% discount

    # Additional method
    def display_info(self):
        super().display_info()
        print("Top Speed:", self.__top_speed, "km/h")


# Objects
b1 = Bike("Honda", "CD70", 120000)
s1 = SportBike("Yamaha", "R1", 2500000, 300)

print("--- Normal Bike ---")
b1.display_info()
print("Discount:", b1.calculate_discount())

print("\n--- Sports Bike ---")
s1.display_info()
print("Discount:", s1.calculate_discount())
```

```python
class Animal:
    def speak(self):
        print("Animal makes sound")

class Dog(Animal):   # Inheritance
    def bark(self):
        print("Dog barks")

d = Dog()
d.speak()   # inherited method
d.bark()    # own method
```
