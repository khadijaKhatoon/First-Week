##  The Student Class System

```python

# STEP 1: Create base class Person with attributes name and age
class Person:
    def __init__(self, name, age):
        self.name = name  # Initialize name attribute
        self.age = age    # Initialize age attribute
    


    # STEP 2: Create method display_info() that prints name and age
    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")



# STEP 3: Create derived class Student that inherits from Person
class Student(Person):


    # STEP 4: Add attributes student_id and major
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)  # Call parent class constructor
        self.student_id = student_id  # Initialize student ID
        self.major = major            # Initialize major
    


    # STEP 5: Override display_info() to print student details
    def display_info(self):
        super().display_info()  # Call parent's display_info method
        print(f"Student ID: {self.student_id}")
        print(f"Major: {self.major}")



# STEP 6: Create derived class GraduateStudent inheriting from Student
class GraduateStudent(Student):


    # STEP 7: Add thesis_topic attribute and override display_info()
    def __init__(self, name, age, student_id, major, thesis_topic):
        super().__init__(name, age, student_id, major)  # Call Student constructor
        self.thesis_topic = thesis_topic  # Initialize thesis topic
    
    # Override display_info() to include thesis topic
    def display_info(self):
        super().display_info()  # Call Student's display_info method
        print(f"Thesis Topic: {self.thesis_topic}")



# STEP 8: Create one object of each class and call display_info()

# Create Person object
person = Person("John Doe", 45)
print("\n--- Person ---")
person.display_info()

# Create Student object
student = Student("Jane Smith", 20, "S12345", "Computer Science")
print("\n--- Student ---")
student.display_info()

# Create GraduateStudent object
grad_student = GraduateStudent("Bob Johnson", 27, "G67890", "Physics", "Quantum Computing")
print("\n--- Graduate Student ---")
grad_student.display_info()
```
