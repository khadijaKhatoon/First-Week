##  The Shopping List Manager 

Build a program that manages a shopping list using lists and dictionaries.

```python
## 1. Create a list named shopping_list containing at least 5 initial items 
##(e.g., "milk", "bread", "eggs", "apples", "chicken").
shopping_list = ["milk", "bread", "eggs", "apples", "chicken"]      ##  List of 5 initial grocery items
print(f"Initial Shopping List: {shopping_list}")      
```

```python
## 2. Print the entire list, and also print the third item in the list.
print(f"The third item is: {shopping_list[2]}")     # Display the complete shopping list 
                                                    # Index 2 = third item
```

```python
## 3. Add two new items to the list using an appropriate list method.
print("Adding 'cheese' and 'coffee'...")  # Inform user about adding items
shopping_list.append("cheese")  # Add "cheese" to the end of the list
shopping_list.append("coffee")  # Add "coffee" to the end of the list
```

```python
## 4. Remove one item from the list.
print("Removing 'chicken'...")  # Remove "chicken" from the list
shopping_list.remove("chicken") # Display the updated list
```

```python
## 5. Create a dictionary named item_prices. The keys should be items from your shopping list (at least 4), 
## and the values should be their prices (as floats).
item_prices = {
    "milk": 2.50,   # Milk price: $2.50
    "bread": 1.99,   # Bread price: $1.99
    "eggs": 3.25,     # Eggs price: $3.25
    "apples": 0.99,     # Apples price: $0.99
    "cheese": 4.50,       # Cheese price: $4.50
    "coffee": 5.99          # Coffee price: $5.99
}
print(f"Item Prices: {item_prices}")
```

```python
## 6. Print the price of one specific item from the dictionary.
specific_item = "bread"   # Choose which item's price to display
print(f"The price of {specific_item} is ${item_prices[specific_item]:.2f}")   # Display specific item's price
```

```python
## 7. Print a formatted summary of your final shopping list and the price dictionary.
print("\n--- Final Summary ---")
print(f"Items to buy: {', '.join(shopping_list)}")  # Display all items to buy
print(f"Price lookup: {item_prices}")   # Show the price dictionary
print("-" * 20)  # Print a line of 20 dashes as footer
```
