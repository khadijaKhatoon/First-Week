### ASSIGNMENT 2: NumPy Array Math Wizard

```python
# Demonstrate broadcasting and universal functions

# Step 1: Import NumPy as np
import numpy as np

print("=" * 60)
print("NUMPY ARRAY MATH WIZARD")
print("=" * 60)


# Step 2: Create arrays
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
arr2 = np.array([10, 20, 30])

print("\n--- Initial Arrays ---")
print(f"arr1 (2x3):\n{arr1}")
print(f"arr2 (1D): {arr2}")



# Step 3: Perform arithmetic operations (demonstrating broadcasting)
print("\n--- Broadcasting Operations ---")


# Explanation of broadcasting (Step 4)
""""
BROADCASTING EXPLANATION:

Broadcasting allows NumPy to perform operations on arrays of different shapes.
Here, arr2 (shape 3,) is automatically stretched to match arr1 (shape 2,3):
- arr2 becomes effectively [[10,20,30], [10,20,30]] for the operation
- This happens without creating actual copies, making it memory efficient
- Rules: Dimensions are compared from right to left, and are compatible if:
  1. They are equal, or
  2. One of them is 1, or
  3. One dimension is missing
"""

# Addition
result_add = arr1 + arr2
print(f"\nAddition (arr1 + arr2):\n{result_add}")

# Subtraction
result_sub = arr1 - arr2
print(f"\nSubtraction (arr1 - arr2):\n{result_sub}")

# Multiplication
result_mul = arr1 * arr2
print(f"\nMultiplication (arr1 * arr2):\n{result_mul}")

# Division
result_div = arr1 / arr2
print(f"\nDivision (arr1 / arr2):\n{result_div}")



# Step 5: Create arr3 and apply universal functions
arr3 = np.array([1, 4, 9, 16, 25])
print("\n" + "=" * 60)
print("UNIVERSAL FUNCTIONS (ufuncs)")
print("=" * 60)
print(f"\narr3: {arr3}")

# Apply np.sqrt() - square root
sqrt_result = np.sqrt(arr3)
print(f"\nnp.sqrt(arr3): {sqrt_result}")

# Apply np.exp() - exponential (e^x)
exp_result = np.exp(arr3)
print(f"np.exp(arr3): {exp_result}")

# Apply np.sin() - sine (in radians)
sin_result = np.sin(arr3)
print(f"np.sin(arr3): {sin_result}")



# Step 6: Create arr4 and demonstrate aggregation functions
arr4 = np.array([[5, 12, 3], [8, 2, 7]])
print("\n" + "=" * 60)
print("AGGREGATION OPERATIONS")
print("=" * 60)
print(f"\narr4:\n{arr4}")

# Sum of all elements
total_sum = np.sum(arr4)
print(f"\nSum of all elements: {total_sum}")

# Max in each column (axis=0 means operate along rows, i.e., down each column)
max_per_column = np.max(arr4, axis=0)
print(f"Max in each column: {max_per_column}")

# Min in each row (axis=1 means operate along columns, i.e., across each row)
min_per_row = np.min(arr4, axis=1)
print(f"Min in each row: {min_per_row}")

# Additional aggregation examples
print(f"\nMean of all elements: {np.mean(arr4):.2f}")
print(f"Standard deviation: {np.std(arr4):.2f}")
print(f"Sum of each row: {np.sum(arr4, axis=1)}")
print(f"Sum of each column: {np.sum(arr4, axis=0)}")



# Step 7: Convert arr1 to float using type casting
print("\n" + "=" * 60)
print("TYPE CASTING")
print("=" * 60)

print(f"\nOriginal arr1 data type: {arr1.dtype}")
print(f"Original arr1:\n{arr1}")

# Method 1: Using astype()
arr1_float = arr1.astype(float)
print(f"\nAfter conversion (astype(float)):")
print(f"New data type: {arr1_float.dtype}")
print(f"arr1 as float:\n{arr1_float}")


# Demonstrate operations with float array
print(f"\nFloat array operations:")
print(f"Square root of float arr1:\n{np.sqrt(arr1_float)}")

print("\n" + "=" * 60)
print("END OF DEMONSTRATION")
print("=" * 60)
```
