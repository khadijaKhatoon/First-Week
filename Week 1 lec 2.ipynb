## Write a script that processes a range of numbers using loops and conditional statements.

```python
## 1. Use a for loop to iterate through numbers from 1 to 20.
#  FizzBuzz from 1 to 20 using a for loop
print("FIZZBUZZ FROM 1 TO 20")
print("-" * 20)

#  Use a for loop to iterate through numbers from 1 to 20
for number in range(1, 21):                  # range(1, 21) generates numbers 1 through 20
    # Step 2: Apply conditional statements to each number
    # Check if number is divisible by both 3 and 5 first (most specific condition)
    if number % 3 == 0 and number % 5 == 0:  # Check divisibility by both 3 and 5
        print("FizzBuzz")                    # Print FizzBuzz if divisible by both
    # Check if number is divisible by 3 only
    elif number % 3 == 0:                    # Check divisibility by 3
        print("Fizz")                        # Print Fizz if divisible by 3
    # Check if number is divisible by 5 only
    elif number % 5 == 0:                    # Check divisibility by 5
        print("Buzz")                        # Print Buzz if divisible by 5
    # If not divisible by 3 or 5, print the number itself
    else:
        print(number)                        # Print the original number

print("\n" + "-" * 30)                       # Print separator line


## 2. For each number:
   ## - If divisible by 3, print "Fizz".
   ## - If divisible by 5, print "Buzz".
   ## - If divisible by both 3 and 5, print "FizzBuzz".
   ## - Otherwise, print the number itself.
#  Countdown based on user input

## 3. After the loop, ask the user for a number.
#  Ask the user for a number
user_number = int(input("Enter a number for countdown: "))  # Get user input and convert to integer


## 5. Use break to stop the countdown if the user's number is greater than 10.
#  Use break to stop if number is greater than 10
if user_number > 10:       # Check if user number exceeds 10
    print(f"Number {user_number} is greater than 10. Countdown stopped.")
else:

## 4. Use a while loop to print a countdown from that number down to 1, and then print "Blast Off!".
    # Use a while loop to print countdown
    count = user_number    # Initialize counter variable with user's number
    while count >= 1:      # Continue loop while count is greater than or equal to 1
        print(f"Countdown: {count}")  # Print current count
        count -= 1         # Decrement count by 1 (same as count = count - 1)
    
    # After loop ends, print Blast Off!
    print("Blast Off!")    # Print blast off message when countdown reaches 0
```
