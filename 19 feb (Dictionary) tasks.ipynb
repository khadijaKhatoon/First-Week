```python
# Dictionary with multiple phone numbers
person = {
    "name": "dija",
    "age": 17,
    "phone_numbers": ["03001234567", "03211234567"] }

print("Original Dictionary:")
print(person)

# Add a new phone number
person["phone_numbers"].append("03451234567")
print("\nAfter Adding a Phone Number:")
print(person)

# Replace first phone number
person["phone_numbers"][0] = "03100000000"
print("\nAfter Replacing First Phone Number:")
print(person)

# Delete second phone number
del person["phone_numbers"][1]
print("\nAfter Deleting Second Phone Number:")
print(person)
```

```python
list1 = [10, 20, 30, 40]
list2 = [1, 2, 3, 4]
print(list1)
print(list2)
addition = [a + b for a, b in zip(list1, list2)]
subtraction = [a - b for a, b in zip(list1, list2)]
multiplication = [a * b for a, b in zip(list1, list2)]
division = [a / b for a, b in zip(list1, list2)]

print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)

list1.append(70)
print("list 1", list1)

```
