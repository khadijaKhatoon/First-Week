```python
import numpy as np

# 1D array
arr1D = np.arange(1, 11)   # [1,2,3,...,10] - 5 se 11 kiya kyuki 1 se 10 tak 10 elements hote hain
print("1D Array:", arr1D)

# --- Indexing & Slicing ---
print("Indexing arr1D[2]:", arr1D[2])        # 3rd element
print("Slicing arr1D[2:7]:", arr1D[2:7])    # elements from index 2 to 6

# --- Convert 1D to 2D ---
arr2D = arr1D.reshape(2, 5)   # 2 rows, 5 columns (10 elements)
print("\nConverted 2D Array:\n", arr2D)

# --- Access element in 2D ---
print("Element at [1,3]:", arr2D[1, 3])   # row=1, col=3

# --- Replace element in 2D ---
arr2D[1, 3] = 99
print("After replacing element at [1,3]:\n", arr2D)

# --- Convert 1D to 3D (shape 5x7) ---
arr1D_large = np.arange(1, 36)   # 35 elements (1 se 35 tak)
arr3D = arr1D_large.reshape(5, 7, 1)   # 5x7 matrix with depth=1 (5*7*1=35)
print("\nConverted 3D Array (5x7x1):\n", arr3D)

# --- Indexing in 3D ---
print("Element at [2,4,0]:", arr3D[2, 4, 0])   # row=2, col=4, depth=0

# --- Convert 3D back to 1D ---
arr1D_back = arr3D.flatten()
print("\nConverted back to 1D:\n", arr1D_back)
```
