### ASSIGNMENT 3: Pandas DataFrame Detective

```python
# Perform basic exploratory data analysis using Pandas


# Step 1: Import pandas as pd and numpy as np
import pandas as pd
import numpy as np

print("=" * 60)
print("PANDAS DATAFRAME DETECTIVE")
print("=" * 60)



# Step 2: Create dictionary with data
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank'],
    'Age': [25, 30, np.nan, 35, 28, np.nan],
    'Score': [85, 92, 78, np.nan, 88, 95],
    'Grade': ['A', 'A', 'B', 'C', 'B', 'A']
}



# Step 3: Create DataFrame df
df = pd.DataFrame(data)

print("\n--- Original DataFrame ---")
print(df)



# Step 4: Basic DataFrame exploration
print("\n" + "=" * 60)
print("BASIC DATAFRAME INFORMATION")
print("=" * 60)

# df.head(3) - Display first 3 rows
print("\n--- First 3 rows (df.head(3)) ---")
print(df.head(3))

# df.shape - Display dimensions
print(f"\n--- DataFrame Shape ---")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# df.columns - Display column names
print(f"\n--- Column Names ---")
print(list(df.columns))

# df.dtypes - Display data types
print(f"\n--- Data Types ---")
print(df.dtypes)



# Step 5: Missing data handling
print("\n" + "=" * 60)
print("MISSING DATA ANALYSIS")
print("=" * 60)

# df.isnull() - Check for missing values
print("\n--- Missing Values (True = missing) ---")
print(df.isnull())

# Summary of missing values
print("\n--- Missing Values Summary ---")
print(df.isnull().sum())

# Fill Age with mean
age_mean = df['Age'].mean()
print(f"\n--- Fill Missing Age with Mean ({age_mean:.2f}) ---")
df_filled = df.copy()  # Create a copy to preserve original
df_filled['Age'] = df_filled['Age'].fillna(age_mean)
print(df_filled)

# Drop rows with missing Score
print("\n--- Drop Rows with Missing Score ---")
df_cleaned = df.dropna(subset=['Score'])  # Drop rows where Score is NaN
print(df_cleaned)



# Step 6: Data selection
print("\n" + "=" * 60)
print("DATA SELECTION")
print("=" * 60)

# Select Name and Score columns
print("\n--- Select Name and Score Columns ---")
name_score = df[['Name', 'Score']]
print(name_score)

# Select rows where Score > 85
print("\n--- Rows where Score > 85 ---")
high_scores = df[df['Score'] > 85]
print(high_scores[['Name', 'Score', 'Grade']])  # Show relevant columns

# Set Name as index
print("\n--- Set Name as Index ---")
df_indexed = df.set_index('Name')
print(df_indexed)



# Step 7: Operations on data
print("\n" + "=" * 60)
print("DATA OPERATIONS")
print("=" * 60)

# Unique grades
print("\n--- Unique Grades ---")
unique_grades = df['Grade'].unique()
print(unique_grades)

# Grade counts
print("\n--- Grade Counts ---")
grade_counts = df['Grade'].value_counts()
print(grade_counts)

# Sort by Score descending
print("\n--- Sort by Score (Descending) ---")
df_sorted = df.sort_values('Score', ascending=False)
print(df_sorted[['Name', 'Score', 'Grade']])

# Additional useful operations
print("\n" + "=" * 60)
print("ADDITIONAL STATISTICS")
print("=" * 60)

# Basic statistics
print("\n--- Statistical Summary ---")
print(df.describe())

# Group by Grade
print("\n--- Average Score by Grade ---")
grade_stats = df.groupby('Grade')['Score'].mean()
print(grade_stats)

# Filter and display final clean DataFrame
print("\n--- Final Clean DataFrame (no missing values) ---")
final_df = df.dropna()  # Drop all rows with any missing values
print(final_df)

print("\n" + "=" * 60)
print("END OF DATAFRAME DETECTIVE")
print("=" * 60)
```
